```python
plt.hist(df["worldwide_gross"])
plt.title("distrubution of worldwide gross")
plt.show()
```

```python
plt.figure(figsize=(15,5))
movie_prodution_budget=df.groupby("movie")["production_budget"].max().sort_values(ascending=True).reset_index().head(30)
sns.barplot(x="movie",y="production_budget",data=movie_prodution_budget)
plt.title("MOVIES PRODUTION BUDGET")
plt.xticks(rotation=90)
plt.show()
```

```python
df[["release_date","movie","genre"]]
```

```python
df[["movie","production_budget","domestic_gross","worldwide_gross"]].head(30)
```

```python
plt.figure(figsize=(14,4))
top_worldwide_gross=df.groupby("movie")["worldwide_gross"].max().sort_values(ascending=False).head(10)
sns.barplot(top_worldwide_gross,label=top_worldwide_gross.index)
plt.title("TOP 10 WORLDWIDE GROSS")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14,4))
top_domestic_gross=df.groupby("movie")["domestic_gross"].max().sort_values(ascending=False).head(10)
sns.barplot(top_domestic_gross,label=top_domestic_gross.index)
plt.title("TOP 10 DOMESTIC GROSS")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14,4))
top_production_budget=df.groupby("movie")["production_budget"].max().sort_values(ascending=False).head(20)
sns.barplot(top_production_budget,label=top_production_budget.index)
plt.title("TOP PRODUTION BUDGET MOVIES LIST")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(14,4))
top_production_budget=df.groupby("movie")["production_budget"].min().sort_values(ascending=True).head(20)
sns.barplot(top_production_budget,label=top_production_budget.index)
plt.title("LESS PRODUTION BUDGET MOVIES LIST")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.scatter(x="production_budget",y="worldwide_gross",data=df,alpha=0.7,color="blue")
plt.title("PRODUCTION BUDGET VS WORLDWIDE GROSS")
plt.grid(True)
plt.legend()
plt.show()
```

```python
sns.violinplot(x="genre",y="worldwide_gross",data=df)
plt.title( "GENRE AND WORLDWIDE GROSS")
plt.show()

```

```python
df.groupby(["release_date"].str.slice(6,8))["worldwide_gross"].mean().plot(kind="line")
plt.title("MEAN FOR WORLDWIDE GROSS OVER A YEAR")
plt.show()
```
